```python
#Calculate the mean value of a field under a categorial data
#In this case, we calculate the mean value of estimated value under every zipcode
df.groupby(['zipcode', 'yearBuilt']).estimated_value.mean()
```

```python
#Cast series to a dataframe
df2 = df.groupby(['zipcode', 'yearBuilt']).estimated_value.mean().reset_index()
```

```python
#Merge multiple dataframes
df3 = pd.merge(df, df2, on='zipcode', how='left')
```
